Download the Pima Indian Diabetes Dataset:

Available from sources like Kaggle.
Contains columns such as:

Pregnancies, Glucose, BloodPressure, SkinThickness, Insulin, BMI,
DiabetesPedigreeFunction, Age, Outcome (whether the patient has
diabetes or not).

Regression Task:

Predict the Blood Pressure of the patients based on other features.
Use Linear Regression model from Scikit-learn.

Classification Task:

Predict whether the patient has diabetes (target column: Outcome).
Use Logistic Regression or K-Nearest Neighbors (KNN) model.

Once Build Evaluate the Model Appropriately.

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import mean_squared_error, r2_score, accuracy_score, classification_report, confusion_matrix
```

```python
# To read csv file for loading dataframe
df = pd.read_csv("/content/drive/MyDrive/Concepts of technologies and AI/diabetes.csv")
```

## Regression Task


```python
# pridict Blood Pressure
X_reg = df.drop(["BloodPressure", "Outcome"], axis = 1)
y_reg = df["BloodPressure"]
X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(X_reg, y_reg, test_size = 0.2, random_state = 42)
```

```python
# linear regression
lin_reg = LinearRegression()
lin_reg.fit(X_train_reg, y_train_reg)
y_pred_reg = lin_reg.predict(X_test_reg)
```

```python
# Evaluating Regression
print("Regression Task:")
print("Mean Squared Error: ", mean_squared_error(y_test_reg, y_pred_reg))
print("R^2 Score: ", r2_score(y_test_reg, y_pred_reg))
```

##Classification Task


```python
# predict Outcome
X_clf = df.drop("Outcome", axis = 1)
y_clf = df["Outcome"]
X_train_clf, X_test_clf, y_train_clf, y_test_clf = train_test_split(X_clf, y_clf, test_size = 0.2, random_state = 42)
```

```python
# Logistic Regression
log_reg = LogisticRegression(max_iter = 200)
log_reg.fit(X_train_clf, y_train_clf)
y_pred_log = log_reg.predict(X_test_clf)
```

```python
# K-Nearest Neighbors
knn = KNeighborsClassifier(n_neighbors = 5)
knn.fit(X_train_clf, y_train_clf)
y_pred_knn = knn.predict(X_test_clf)
```

```python
# Evaluating Classification
print("Classification Task:")
print("Logistic Regression Accuracy:", accuracy_score(y_test_clf, y_pred_log))
print("KNN Accuracy:", accuracy_score(y_test_clf, y_pred_knn))
print("\nLogistic Regression Classification Report:\n", classification_report(y_test_clf, y_pred_log))
print("KNN Classification Report:\n", classification_report(y_test_clf, y_pred_knn))
print("Logistic Regression Confusion Matrix:\n", confusion_matrix(y_test_clf, y_pred_log))
print("KNN Confusion Matrix:\n", confusion_matrix(y_test_clf, y_pred_knn))
```
