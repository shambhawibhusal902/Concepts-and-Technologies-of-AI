Task - 3 - Outlier Detection:

Tasks:
  
  1. Identify outlier countries in South Asia based on their Score and GDP per Capita.
  
  2. Define outliers using the 1.5 Ã— IQR rule.
  
  3. Create a scatter plot with GDP per Capita on the x-axis and Score on the   y-axis, highlighting outliers in a different color.
  4. Discuss the characteristics of these outliers and their potential impact on regional averages

```python
# 2.
# defining outliers using the 1.5 * IQR rule
# calculating IQR for score
Q1_score = south_asia_dataset['score'].quantile(0.25)
Q3_score = south_asia_dataset['score'].quantile(0.75)
IQR_score = Q3_score - Q1_score

# calculating IQR for score
Q1_gdp = south_asia_dataset['Log GDP per capita'].quantile(0.25)
Q3_gdp =south_asia_dataset['Log GDP per capita'].quantile(0.75)
IQR_gdp = Q3_gdp - Q1_gdp

# calculating upper bound and lower bound for outliers for gdp
lower_bound_gdp = Q1_gdp - 1.5 * IQR_gdp
upper_bound_gdp = Q3_gdp + 1.5 * IQR_gdp

# calculating upper bound and lower bound for outliers for score
lower_bound_score = Q1_score - 1.5 * IQR_score
upper_bound_score = Q3_score + 1.5 * IQR_score

```

```python
# 1.
# identifying outliers
outliers = south_asia_dataset[(south_asia_dataset['score'] < lower_bound_score) | (south_asia_dataset['score'] > upper_bound_score) | (south_asia_dataset['Log GDP per capita'] < lower_bound_gdp) | (south_asia_dataset['Log GDP per capita'] > upper_bound_gdp)]
# displaying outliers
outliers
```

```python
# 3.
# creating scatter plot with GDP per Capita on the x-axis and Score on the y-axis
plt.figure(figsize = (10, 6))
plt.scatter(south_asia_dataset['Log GDP per capita'], south_asia_dataset['score'], color = 'skyblue', label = 'Normal')
plt.scatter(outliers['Log GDP per capita'], outliers['score'], color ='green', label = 'outliers')
plt.xlabel('Log GDP per capita')
plt.ylabel('score')
plt.title('Relation between Log GDP per capita and score(highlighting outliers)')
plt.legend()

# displaying scatter plot
plt.show()
```

4.
Outliers are the data points which are significantly different from other observations.

Characterstics of Outliers in South asia are:-
1. High GDP per capita but low score.
2. Low GDP per capita but high score.

Their potential impact on regional averages are:-
1. Skewing Averages
2. Misleading Insights
3. Highlighting Unique cases
